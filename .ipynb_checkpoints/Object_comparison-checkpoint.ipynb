```python
# Remove all houses with an area of less than 25000
for numb in houses_df.House_number:
    x_min,x_max,y_min,y_max = list(houses_df[houses_df['House_number'] == numb]['[x_min,x_max,y_min,y_max]'].values)[0]
    if (x_max - x_min)*(y_max - y_min) <= 25000:
        houses_df.drop(houses_df[houses_df['House_number'] == numb].index[0],inplace = True)
```

```python
# Average probability score
mean_prob_score = []
for house in houses_df['House_number']:
    x_min,x_max,y_min,y_max = list(houses_df[houses_df['House_number'] == house]['[x_min,x_max,y_min,y_max]'].values)[0]
    mean_prob_score.append([sigmoid(prob_ret[y_min:y_max,x_min:x_max]).mean(), house])
```

```python
# Average variance score
mean_var_score = []
for house in houses_df['House_number']:
    x_min,x_max,y_min,y_max = list(houses_df[houses_df['House_number'] == house]['[x_min,x_max,y_min,y_max]'].values[0])
    mean_var_score.append([np.var(sigmoid(prob_ret[y_min:y_max,x_min:x_max])), house])
```

```python
set_ = np.array(list(zip(np.array(mean_var_score)[:,0], np.array(mean_prob_score)[:,0])))
#set_ = preprocessing.scale(set_)
#print(set_.shape)
```

```python
from pycpd import RigidRegistration, DeformableRegistration

def visualize(iteration, error, X, Y, ax):
    plt.cla()
    ax.scatter(X[:, 0],  X[:, 1], color='red', label='Target')
    ax.scatter(Y[:, 0],  Y[:, 1], color='blue', label='Source')
    plt.text(0.87, 0.92, 'Iteration: {:d}\nQ: {:06.4f}'.format(
        iteration, error), horizontalalignment='center', verticalalignment='center', transform=ax.transAxes, fontsize='x-large')
    ax.legend(loc='upper left', fontsize='x-large')
    plt.draw()
        
def main(X,Y):

    fig = plt.figure()
    fig.add_axes([0, 0, 1, 1])
    callback = partial(visualize, ax=fig.axes[0])
    #reg = DeformableRegistration(**{'X': X, 'Y': Y},alpha = 7, beta = 10, max_iterations = 50)
    reg = RigidRegistration(**{'X': X, 'Y': Y},max_iterations = 50)
    #reg = AffineRegistration(**{'X': X, 'Y': Y},max_iterations = 300)
    reg.register(callback)
    params = reg.get_registration_parameters()
    Y_1 = reg.register(callback)
    plt.show()
    return(params)
```

```python
def get_edge(image,color):
    lower = np.array(color)
    upper = np.array(color)

    img = np.copy(image)

    thresh = cv.inRange(img, lower, upper)
    th = np.array(list(zip(np.where(thresh == 255)[0],np.where(thresh == 255)[1])))
    contours, hierarchy = cv.findContours( thresh.copy(), cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
    cv.drawContours(img, contours, -1, 133, 1)
    
    l = np.array(133)
    u = np.array(133)

    thresh = cv.inRange(img, l, u)
    edge = np.array(list(zip(np.where(thresh == 255)[0],np.where(thresh == 255)[1])))
    return(edge)
```
